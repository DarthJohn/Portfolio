### 2.2.2 Пропуски в столбце year_of_release

Проанализируем пропуски в данном столбце

```python
df[df['year_of_release'].isna()].head(20)
```

Какой-то особой системы в пропусках года выхода игр не прослеживается. Это игры разных платформ, разных жанров, разных продаж и рейтингов. 

Пропуски, скорее всего, возникли из-за повреждения данных, из-за человеческого фактора или при объединении датасетов в одном из них содержались данные не по всем играм.

Предположим, что пропуски в столбце year_of_release можно заменить на год, указанный в названии игры, например FIFA Soccer 2004, возможно, вышла в 2004 году. Проверим эту гипотезу.

```python
def year_of_game(index):
    for year in range(1980, 2017):
        if str(year) in index['name']:
            print(f" {index['name']} \t {index['year_of_release']}")
```

```python
df.apply(year_of_game, axis=1)
```

Результаты получаются довольно разными:
- часть игр вышла в том году, который указан у них в названии, например, UFC 2009 Undisputed вышла в 2009 году;
- часть игр вышла за год до числа, указанного в названии, например, NBA Live 2002 вышла в 2001 году;
- есть игры, для которых год указанный в названии не имеет ничего общего с годом выхода, например SimCity 2000 вышла в 1992 году.

Более того по данным Википедии FIFA Soccer 2004 вышла 24 октября 2003. Гипотеза не подтвердилась.

Получается, что автоматизировать заполнение пропусков года выхода игр по этому признаку не получится. Других способов заполнить пропуски в этом столбце нет. Поэтому оставим их незаполненными.

### 2.2.3 Пропуски в столбце critic_score

Проанализируем пропуски в данном столбце

```python
df[(df['critic_score'].isna())].head(10)
```

По первым 10 строкам бросается в глаза, что для всех игр пропуски есть одновременно и в столбце critic_score, и в user_score, и в rating. Возможно здесь дело в том, что анализируемый датафрейм получен объединением нескольких датафреймов и тот, в котором содержались оценки и рейтинги, просто не содержит данные о некоторых играх, которые есть в других датафреймах.

Кроме того, мы видим тут, что оценок нет, в том числе, у игр 80х и 90х годов, а также у не самых известных игр, например Nintendogs. Возможно, что для старых, а также не слишком известных игр просто нет оценок критиков.

Теперь выясним есть ли строки, в которых пропуски только в столбце critic_score.

```python
df[(df['critic_score'].isna())&(~df['user_score'].isna())&(~df['rating'].isna())].head(10)
```

Да, такие строки есть. На примере пропусков для двух версий Call of Duty: Advanced Warfare для PS3 и X360 посмотрим есть ли оценки критиков для этой игры на других платформах.

```python
df[df['name'] == 'Call of Duty: Advanced Warfare']
```

Оказывается, что есть. Вполне логично, что критики писали свои рецензии для актуальных консолей, а для устаревающих на момент выхода игры – нет. Пользователи же оценили игру на всех платформах.

Заполнить пропуски для рейтинга критиков возможно для мультиплатформенных игр, однако рейтинг может очень сильно колебаться в зависимости от платформы. Это видно по пользовательским оценкам для Call of Duty: Advanced Warfare.

Любая попытка заполнить здесь пропуски приведёт к искажению данных, поэтому заполнять их мы не будем.

### 2.2.4 Пропуски в столбце user_score

### 2.2.5 Пропуски в столбце rating

Проанализируем пропуски в данном столбце для тех строк, где нет пропусков в critic_score и rating.

```python
df[(df['rating'].isna())&(~df['critic_score'].isna())&(~df['user_score'].isna())].head(10)
```

Можно предположить, что каким-то играм рейтинги просто не выдаются, либо же техническую ошибку, из-за которой для части игр отсутствуют возрастные рейтинги.

На примере Wolfenstein: The Old Blood, игры у которой просто не может не быть возрастного рейтинга, проверим его наличие у версий для разных платформ.

```python
df[df['name'] == 'Wolfenstein: The Old Blood']
```

```python
df['rating'].mode()
```

Видим, что версия для PC имеет рейтинг M, а версии для PS4 и XOne только пропуск.

Очень маловероятно, чтобы возрастные рейтинги для одной и той же игры на разных платформах были разные, поэтому заполним пропуски в таких случаях возрастным рейтингом той версии игры, информация о которой есть.

Число пропусков в столбце rating до обработки.

```python
df['rating'].isna().sum()
```

```python
# цикл проходит по всем строкам, у которых есть дубликат в столбце name и нет пропуска в столбце rating 
for i in df[(df['name'].duplicated())&(~df['rating'].isna())].index:
    # для всех игр, с тем же именем, что у игры в текущей строке проставляется рейтинг этой игры
    df.loc[df['name'] == df.loc[i]['name'], 'rating'] =  df.loc[i]['rating']
```

Число пропусков в столбце rating после обработки.

```python
df['rating'].isna().sum()
```

Заполнили 316 пропусков.

Проверим заполнились про пропуски для Wolfenstein: The Old Blood

```python
df[df['name'] == 'Wolfenstein: The Old Blood']
```

Заполнилось всё правильно.

Остальные пропуски в столбце rating мы заполнить не можем.

# 5 Статистический анализ

Проверим несколько гипотез

## 5.1 Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Формулируем нулевую гипотезу:

**Средние пользовательские рейтинги платформ Xbox One и PC равны.**

Формулируем альтернативную гипотезу:

**Средние пользовательские рейтинги платформ Xbox One и PC не равны.**

Критерием для проверки гипотезы будет является уровень статистической значимости 5%. Если величина p-value будет меньше уровня статистической значимости, то это значит, что при верности нулевой гипотезы (равенстве средних рейтингов для платформ Xbox One и PC), вероятность получить выборки пользовательских рейтингов, представленные в наших данных, меньше 5%.

```python
result = st.ttest_ind(df_actual[df_actual['platform'] == 'PC']['user_score'].dropna(), \
                      df_actual[df_actual['platform'] == 'XOne']['user_score'].dropna(), \
                      equal_var=False)
alpha = 0.05

print(result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

## 5.2 Средние пользовательские рейтинги жанров Action и Sports разные

Формулируем нулевую гипотезу:

**Средние пользовательские рейтинги жанров Action и Sports не равны.**

Формулируем альтернативную гипотезу:

**Средние пользовательские рейтинги жанров Action и Sports равны.**

Критерием для проверки гипотезы будет является уровень статистической значимости 5%. Если величина p-value будет больше уровня статистической значимости, то это значит, что при верности нулевой гипотезы (неравенстве средних рейтингов Action и Sports), вероятность получить выборки пользовательских рейтингов, представленные в наших данных, меньше 5%.

```python
result = st.ttest_ind(df_actual[df_actual['genre'] == 'Action']['user_score'].dropna(), \
                      df_actual[df_actual['genre'] == 'Sports']['user_score'].dropna(), \
                      equal_var=False)
alpha = 0.05

print(result.pvalue)

if result.pvalue > alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
